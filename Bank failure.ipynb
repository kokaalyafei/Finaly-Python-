```python
import plotly.plotly as py
import pandas as pd

df = pd.read_csv('Banklist.States.csv')

for col in df.columns:
    df[col] = df[col].astype(str)

scl = [[0.0, 'rgb(242,240,247)'],[0.2, 'rgb(218,218,235)'],[0.4, 'rgb(188,189,220)'],            [0.6, 'rgb(158,154,200)'],[0.8, 'rgb(117,107,177)'],[1.0, 'rgb(84,39,143)']]

df['text'] = df['States'] + '<br>' +    'Failure Count '+df['FailuresByStates']

data = [ dict(
        type='choropleth',
        colorscale = scl,
        autocolorscale = False,
        locations = df['States'],
        z = df['FailuresByStates'].astype(float),
        locationmode = 'USA-states',
        text = df['text'],
        marker = dict(
            line = dict (
                color = 'rgb(255,255,255)',
                width = 2
            ) ),
        colorbar = dict(
            title = "Bank Failures")
        ) ]

layout = dict(
        title = 'United States Bank Failure from 2000-2018<br>(Hover for breakdown)',
        geo = dict(
            scope='usa',
            projection=dict( type='albers usa' ),
            showlakes = True,
            lakecolor = 'rgb(255, 255, 255)'),
             )
    
fig = dict( data=data, layout=layout )
py.iplot( fig, filename='d3-cloropleth-map' )
```

```python
import pandas as pd
import folium
 

state_geo = 'C:\\Users\\kokaa\\Desktop\\us-states.json'
 
State_data = pd.read_csv('US_Unemployment_Oct2012.csv')
 
m = folium.Map(location=[37, -102], zoom_start=5)
 

m.choropleth(
 geo_data=state_geo,
 name='choropleth',
 data=state_data,
 columns=['States', 'Failure_Count'],
 key_on='feature.id',
 fill_color='YlGn',
 fill_opacity=0.7,
 line_opacity=0.2,
 legend_name='Failure_Count'
)
folium.LayerControl().add_to(m)
 
# Save to html
m.save('StateBank.html')
```
